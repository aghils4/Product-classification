```python
import numpy as np
import pandas as pd
```

```python
dtrain=pd.read_csv('train.csv')
dtest=pd.read_csv('test.csv')
```

```python
len(dtrain['Sub-Industry'].unique())
```

```python
dtest
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
stopwords=[]
```

```python
nltk_stopwords=nltk.corpus.stopwords.words('english')
stopwords.extend(nltk_stopwords)
```

```python
with open ('stop') as stopwords_file:
    st=stopwords_file.readlines()
stopwords1=[line.replace('\n','')for line in st]
stopwords1.extend(stopwords)
```

```python
stopwords1
```

```python
from nltk.stem.snowball import SnowballStemmer
nltk.download('punkt')
```

```python
stemmer = SnowballStemmer("english")
```

```python
dataset=pd.DataFrame(columns=('title_body','Sub_Industry'))
for index,row in dtrain.iterrows():
    title_body=row['Ticker']+' '+row['Company Name']+' '+row['Sector']+' '+row['Industry Group']+' '+row['Industry']+' '+row['Description']
    title_body_tok=nltk.word_tokenize(title_body)
    title_body_tok_filtered=[w for w in title_body_tok if not w in stopwords1]
    title_body_tok_filtered_stem= [stemmer.stem(w) for w in title_body_tok_filtered]
    dataset.loc[index]={
        'title_body': ' '.join(title_body_tok_filtered_stem),
        'Sub_Industry': row['Sub-Industry']
    }
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector=TfidfVectorizer()
vector.fit(dataset['title_body'])
```

```python
x=vector.transform(dataset['title_body'])
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(dataset['Sub_Industry'])
```

```python
y=le.transform(dataset['Sub_Industry'])
```

```python
len(le.classes_)
```

```python
len(y)
```

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train , X_test,Y_train , Y_test=train_test_split(x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=50)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
dataset1=pd.DataFrame(columns=('title_body','a'))
for index,row in dtest.iterrows():
    title_body=row['Ticker']+' '+row['Company Name']+' '+row['Sector']+' '+row['Industry Group']+' '+row['Industry']+' '+row['Description']
    title_body_tok=nltk.word_tokenize(title_body)
    title_body_tok_filtered=[w for w in title_body_tok if not w in stopwords1]
    title_body_tok_filtered_stem= [stemmer.stem(w) for w in title_body_tok_filtered]
    dataset1.loc[index]={
        'title_body': ' '.join(title_body_tok_filtered_stem),
        'a': row['Ticker']
    }
```

```python
x1=vector.transform(dataset1['title_body'])
```
